1.2 Calculate the minimum, maximum, mean price for the column “Close Price” using pandas library.

```python
import pandas as pd
import numpy as np
# import the csv file into dataframe.
#Start Code
dataset = pd.read_csv('IDBI.csv')
#End code
# Find the maximum, minimum and mean for the column “Close Price”
#Start Code
Maximum = dataset['Close Price'].max()
Minimum = dataset['Close Price'].min()
Mean = dataset['Close Price'].mean()
#End Code
print('Mean :',Mean)
print('Minimum:', Minimum)
print('Maximum:”', Maximum)
```

1.3 Add a column "Day Perc Change where the values are the daily change in percentage i.e. the percentage change between 2 consecutive days closing prices. Instead of using the basic mathematical formula for computing the same, use 'pct_change()' function provided by Pandas for dataframes. You will note that the first entry of the column will have a 'Nan' value. Remove the first row.


```python
dataset = dataset.assign(Day_perc_change=np.nan)
#Start code
dataset.Day_perc_change =dataset['Close Price'].pct_change()
dataset=(dataset.drop(dataset.index[0]))
print(dataset)
#End code
```


1.4 Convert the Date column into the type datetime64 and plot the closing price of each of the days for the entire time frame to get an idea of what the general outlook of the stock is.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dataset.Date=pd.to_datetime(dataset.Date)
plt.style.use('seaborn')
#write the code for plotting
#Start code
print(dataset)
plt.figure(figsize=(16,8))
dataset.index=dataset['Date']
plt.plot(dataset['Close Price'], label='Close Price history')
#End Code
```
